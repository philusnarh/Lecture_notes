The descriptive statistics of the physico-chemical measurements across the three regions (Central, Greater Accra, and Western) during the dry and wet seasons reveal significant seasonal and regional variations in water quality.

Starting with **pH levels**, the Central Region shows relatively stable pH values across both seasons, with an average of 7.94 in the dry season and 8.00 in the wet season, reflecting minor seasonal shifts. The Greater Accra Region exhibits a higher pH mean during the wet season (8.33) compared to the dry season (7.99), suggesting a potential influence of seasonal rainfall on water alkalinity. The Western Region stands out for having a lower pH mean in the dry season (6.96) compared to the wet season (7.46), indicating an increase in acidity during drier periods, which could be attributed to reduced water flow or pollution concentrations.

For **conductivity (Cond.)**, a measurement that reflects the water's ability to conduct electrical current and thus indicates the presence of dissolved ions, there is a clear distinction between the regions. The Central Region shows the highest average conductivity, with means of 12.54 µS/cm (dry) and 12.75 µS/cm (wet), suggesting consistent ion concentrations throughout the year. Greater Accra, on the other hand, presents lower conductivity, especially in the wet season (10.07 µS/cm), indicating dilution effects from rainfall. The Western Region has higher variability, with conductivity rising significantly in the wet season (12.75 µS/cm dry vs. 11.96 µS/cm wet), suggesting potential contamination sources or mineral deposits that may leach into the water during wetter periods.

In terms of **total dissolved solids (TDS)**, which represent the concentration of dissolved substances in water, the Central and Western Regions show higher averages in the dry season, with values of 6.51 ppm and 6.13 ppm, respectively, while Greater Accra experiences lower TDS levels in both seasons, particularly in the wet season (4.72 ppm). This could indicate that rainfall in Greater Accra dilutes the water more effectively, reducing dissolved solids, whereas the Central and Western Regions have more consistent or higher levels of dissolved substances.

The **temperature** of the water samples exhibits a slight decrease in the wet season across all regions, which is consistent with the cooling effects of rain. The Central Region's temperature ranges from an average of 22.60°C (dry) to 22.09°C (wet), while Greater Accra shows a small rise in the wet season (23.04°C compared to 22.18°C dry). The Western Region's temperatures also drop slightly during the wet season (23.03°C dry to 22.39°C wet), possibly due to cloud cover and rain cooling the surface water.

**Turbidity**, which measures water clarity, shows low and relatively stable values across both seasons and regions, indicating generally clear water samples. However, the Western Region exhibits slightly higher turbidity during the dry season (mean of 0.03 NTU) compared to the wet season (0.02 NTU), suggesting that sediment or suspended particles may be more concentrated in drier periods.

For **alkalinity**, there is little seasonal variation across the regions, but slight differences are observed. In the Central Region, alkalinity is slightly higher during the dry season (7.90 mg/L) compared to the wet season (7.44 mg/L), indicating minor changes in the water’s buffering capacity. The Greater Accra and Western Regions also exhibit small variations, with alkalinity remaining relatively stable across seasons, suggesting that the water maintains its ability to neutralize acids effectively, regardless of rainfall.

The **chemical oxygen demand (COD)**, which reflects the amount of organic matter in the water, reveals regional and seasonal fluctuations. The Central and Western Regions show higher COD values in the dry season (7.72 mg/L and 8.23 mg/L, respectively), which may indicate that organic pollutants are more concentrated during periods of reduced water flow. Conversely, Greater Accra presents lower COD values, particularly during the wet season (4.97 mg/L), suggesting a dilution of organic pollutants due to increased water volume.

Finally, **biochemical oxygen demand (BOD)**, a measure of the amount of oxygen required by aerobic microorganisms to decompose organic material, also demonstrates slight regional and seasonal differences. In the Central and Greater Accra Regions, the BOD values are relatively low, but the wet season in the Central Region sees a notable increase (1.67 mg/L compared to 1.18 mg/L in the dry season), indicating higher levels of biodegradable organic matter during periods of increased water flow. The Western Region also shows a similar trend, with BOD values rising during the wet season, reflecting the potential influence of runoff and organic matter accumulation.

In summary, the physico-chemical characteristics of water in these regions reveal seasonal shifts, with certain parameters like pH, conductivity, TDS, and COD being more influenced by rainfall, while others, like temperature and turbidity, remain more stable. The differences between regions also highlight the varying impacts of geographical and environmental factors on water quality.

## Comparative Analysis: Seasons and Regions

````python
import pandas as pd
import scipy.stats as stats

# Separate data into dry and wet seasons for a parameter (e.g., pH)
dry = df[df['Season'] == 'Dry']
wet = df[df['Season'] == 'Wet']

# Paired t-test for normally distributed data (e.g., pH)
stat, p_value_ttest = stats.ttest_rel(dry['pH'], wet['pH'])
print(f'Paired t-test: stat={stat:.3f}, p-value={p_value_ttest:.3f}')

# Wilcoxon Signed-Rank Test for non-normally distributed data (e.g., TDS)
stat, p_value_wilcoxon = stats.wilcoxon(dry['TDS (ppm)'], wet['TDS (ppm)'])
print(f'Wilcoxon Signed-Rank Test: stat={stat:.3f}, p-value={p_value_wilcoxon:.3f}')
````

````python
import pandas as pd
import scipy.stats as stats

# Define parameters to test
parameters = ['pH', 'TDS (ppm)', 'Tempt (°C)', 'COD (mg/L)', 'BOD (mg/L)',
              'Turbidity (NTU)', 'Alkalinity (mg/L)', 'Cond. (µ/cm)']

# Prepare a results DataFrame
results = pd.DataFrame(columns=['Variable',
                                 'Shapiro Stat', 'Shapiro p-val',
                                 'Anderson Stat', 'Anderson Normal?',
                                 'KS Stat', 'KS p-val',
                                 'Paired t-test Stat', 'Paired t-test p-val',
                                 'Wilcoxon Stat', 'Wilcoxon p-val'])

# Iterate through parameters
for param in parameters:
    # Get dry and wet season data
    dry_data = dry[param].dropna()
    wet_data = wet[param].dropna()

    # Perform normality tests
    shapiro_stat, shapiro_pval = stats.shapiro(dry_data)
    anderson_result = stats.anderson(dry_data)
    ks_stat, ks_pval = stats.kstest(dry_data, 'norm', args=(dry_data.mean(), dry_data.std()))

    # Check if Anderson-Darling test statistic indicates normality
    critical_values = anderson_result.critical_values
    significance_level = anderson_result.significance_level

    # Determine if the data is normal based on the critical value for the 5% significance level
    anderson_normal = any(anderson_result.statistic < cv for cv in critical_values)

    # Determine paired t-test or Wilcoxon test
    if shapiro_pval > 0.05:  # Normally distributed
        t_stat, t_pval = stats.ttest_rel(dry_data, wet_data)
        wilcoxon_stat, wilcoxon_pval = None, None  # Not applicable
    else:  # Not normally distributed
        t_stat, t_pval = None, None  # Not applicable
        wilcoxon_stat, wilcoxon_pval = stats.wilcoxon(dry_data, wet_data)

    # Create a new DataFrame for the results of the current parameter
    param_results = pd.DataFrame({
        'Variable': [param],
        'Shapiro Stat': [shapiro_stat],
        'Shapiro p-val': [shapiro_pval],
        'Anderson Stat': [anderson_result.statistic],
        'Anderson Normal?': [anderson_normal],  # True if normal, False otherwise
        'KS Stat': [ks_stat],
        'KS p-val': [ks_pval],
        'Paired t-test Stat': [t_stat],
        'Paired t-test p-val': [t_pval],
        'Wilcoxon Stat': [wilcoxon_stat],
        'Wilcoxon p-val': [wilcoxon_pval]
    })

    # Concatenate the results
    results = pd.concat([results, param_results], ignore_index=True)

# Display results
results
````

The statistical table presents a comprehensive analysis comparing the dry and wet seasons across three regions, focusing on key physico-chemical parameters like pH, TDS, Temperature (Tempt), COD, BOD, Turbidity, Alkalinity, and Conductivity. Several statistical tests were performed to assess normality and differences, including the Shapiro-Wilk test, Anderson-Darling test, Kolmogorov-Smirnov (KS) test, paired t-tests, and Wilcoxon tests.

### pH:
The Shapiro-Wilk test for **pH** shows a statistic of 0.850, with a p-value of 0.0043, indicating that the data significantly deviates from normality. This is confirmed by the Anderson-Darling test, with a statistic of 1.214 and a determination that the data is not normal. The KS test shows a borderline value with a statistic of 0.240 and a p-value of 0.149, which is not statistically significant. For the paired t-test, there are no results due to missing values (NaN), but the Wilcoxon test shows a significant result (statistic = 54.5, p = 0.0319), implying that there are significant differences in pH between the seasons or regions.

### TDS (Total Dissolved Solids):
For **TDS (ppm)**, the Shapiro-Wilk test (stat = 0.958, p = 0.471) and Anderson-Darling test (stat = 0.294, normal) suggest that the data follows a normal distribution. The KS test shows no significant deviation from the reference distribution (stat = 0.122, p = 0.875). The paired t-test does not find a statistically significant difference (t = -0.0768, p = 0.939), and the Wilcoxon test was not performed for this parameter. Overall, TDS does not exhibit significant seasonal or regional differences.

### Temperature (Tempt °C):
**Temperature** shows no significant deviation from normality, as indicated by the Shapiro-Wilk (stat = 0.934, p = 0.162) and Anderson-Darling (stat = 0.455, normal) tests. The KS test also shows no significant deviation (stat = 0.153, p = 0.655). The paired t-test also fails to find significant differences (t = 0.1555, p = 0.878). No Wilcoxon test was conducted. Therefore, there are no significant differences in temperature between the seasons or regions.

### COD (Chemical Oxygen Demand):
The **COD** parameter is normally distributed according to both the Shapiro-Wilk (stat = 0.970, p = 0.730) and Anderson-Darling (stat = 0.223, normal) tests. The KS test also indicates no significant differences (stat = 0.0978, p = 0.976). While the paired t-test shows a positive trend (t = 1.5458), the p-value (0.1378) is not low enough to confirm statistical significance. Overall, the results suggest that COD does not vary significantly between seasons or regions.

### BOD (Biochemical Oxygen Demand):
**BOD** shows a clear deviation from normality, as indicated by the Shapiro-Wilk test (stat = 0.537, p = 4.6e-07) and the Anderson-Darling test (stat = 4.878, non-normal). The KS test confirms these findings (stat = 0.463, p = 0.00012), indicating significant deviations from the reference distribution. The paired t-test is not performed due to missing values, but the Wilcoxon test shows a marginally non-significant result (stat = 62.0, p = 0.0647), suggesting some differences between the dry and wet seasons or among the regions, though the evidence is not strong enough to confirm this definitively.

### Turbidity:
**Turbidity** also fails normality tests, with the Shapiro-Wilk test (stat = 0.793, p = 0.0005) and Anderson-Darling test (stat = 1.801, non-normal) both rejecting normality. The KS test shows a significant result (stat = 0.317, p = 0.0227), suggesting that turbidity significantly deviates from the reference distribution. The Wilcoxon test (stat = 60.5, p = 0.6955) indicates no significant differences across seasons or regions. Therefore, despite non-normality, the data do not show significant differences in turbidity.

### Alkalinity:
For **Alkalinity**, both the Shapiro-Wilk (stat = 0.852, p = 0.0046) and Anderson-Darling (stat = 1.162, non-normal) tests reject the assumption of normality. The KS test, however, does not indicate significant differences (stat = 0.218, p = 0.236). No paired t-test results are available, and the Wilcoxon test (stat = 96.0, p = 0.5168) also shows no significant differences.
````python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multicomp import MultiComparison

# List of parameters to analyze
parameters = ['pH', 'TDS (ppm)', 'Tempt (°C)', 'COD (mg/L)', 'BOD (mg/L)',
              'Turbidity (NTU)', 'Alkalinity (mg/L)', 'Cond. (µ/cm)']

# Prepare results storage
results_list = []

for param in parameters:
    # Prepare data for Kruskal-Wallis Test
    group_data = [group[param].dropna() for name, group in df.groupby('Region')]

    # Perform Kruskal-Wallis Test
    kruskal_stat, p_val = stats.kruskal(*group_data)

    # Store the results in a dictionary
    results_list.append({
        'Variable': param,
        'Kruskal Stat': kruskal_stat,
        'p-value': p_val
    })

    # If Kruskal-Wallis is significant, perform Dunn's post-hoc test
    if p_val < 0.05:
        mc = MultiComparison(df[param], df['Region'])
        posthoc_results = mc.tukeyhsd()
        print(f"\nTukey's HSD results for {param}:\n", posthoc_results)

# Convert the results list to a DataFrame
kruskal_results = pd.DataFrame(results_list)

# Display Kruskal-Wallis results
print("\nKruskal-Wallis Results:")
kruskal_results
````

The **Elbow Method** is a technique used in K-means clustering to determine the optimal number of clusters (K) for a given dataset. The method is based on the idea of plotting the **within-cluster sum of squares (WCSS)**, which represents the sum of squared distances between each point and the centroid of its assigned cluster, against the number of clusters (K).

### Steps in the Elbow Method:
1. **Run K-means clustering** for different values of K (typically starting from K=1 up to a reasonable maximum, e.g., K=10).
2. For each value of K, calculate the **WCSS**.
3. **Plot the WCSS values** against the corresponding number of clusters.
4. The plot generally forms a curve with a clear decline in WCSS as K increases.
5. **Find the "elbow" point** on the curve, which is the value of K after which the rate of decline slows down significantly. This point suggests the optimal number of clusters because increasing K beyond this point does not provide substantial improvement in reducing WCSS.

### Explanation of the "Elbow":
- The elbow point represents a balance between **variance reduction** within clusters and **model complexity**.
- Before the elbow, adding more clusters significantly reduces WCSS, meaning that the clusters are becoming more refined and better at grouping similar data points.
- After the elbow point, adding more clusters results in only small reductions in WCSS, indicating diminishing returns and potentially overfitting the data.

In summary, the **Elbow Method** helps to find a good trade-off between fewer clusters (underfitting) and too many clusters (overfitting), leading to a more meaningful clustering solution.
