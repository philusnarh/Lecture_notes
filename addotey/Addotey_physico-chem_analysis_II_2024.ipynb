The physico-chemical analysis of tap water samples from the Central, Greater Accra, and Western regions during both dry and wet seasons provides insights into water quality variations. Here, the key parameters are discussed to establish trends, compare regions, and evaluate seasonal impacts.

### **pH**
The pH levels across all regions are within acceptable drinking water standards (6.5–8.5), although distinct variations exist. Central region water shows higher mean pH values in both seasons, especially during the wet season (8.45). This indicates alkaline water, likely influenced by local geology or chemical inputs. Greater Accra and Western regions also exhibit slightly alkaline pH, but Western region values are closer to neutrality, with a mean of 6.96 in the dry season. Seasonal shifts are evident, with higher pH values generally observed during wet seasons across all regions, suggesting possible dilution or changes in water chemistry due to runoff.

### **Electrical Conductivity (Cond.)**
Electrical conductivity reflects the ion concentration in the water. The Central region shows significantly higher mean conductivity (15.32 µS/cm in the dry season), indicating greater dissolved solids, possibly from geological sources or anthropogenic activities. Conversely, Greater Accra records lower conductivity (10.43 µS/cm in the dry season), suggesting comparatively lower ion presence. Interestingly, conductivity in all regions tends to decrease during the wet season, possibly due to dilution effects from rainwater.

### **Total Dissolved Solids (TDS)**
The Central region has the highest TDS levels in the dry season (mean 7.7 ppm), aligning with its high conductivity, as TDS is directly correlated with ion content. The Western and Greater Accra regions show similar trends, with TDS slightly lower during the wet season, indicating seasonal dilution. However, all regions remain well below the WHO-recommended TDS limit of 500 ppm for drinking water.

### **Temperature**
Water temperature varies significantly between seasons and regions, influenced by climatic conditions. The Central region exhibits the widest range, with a dry season mean of 20.18°C and a wet season mean of 23.03°C. This trend of higher wet season temperatures is consistent across all regions and reflects ambient temperature differences, which can affect water solubility and biological activity.

### **Turbidity**
Turbidity levels across all regions are remarkably low, with averages below 0.05 NTU, suggesting excellent water clarity. The slight increase in turbidity during wet seasons, notably in Greater Accra, may indicate the influence of sediment-laden runoff. However, the levels remain far below the WHO limit of 5 NTU for drinking water.

### **Alkalinity**
Alkalinity, which measures water's capacity to neutralize acids, shows relatively uniform values across regions, with means around 7–8 mg/L. A marginal increase during the wet season in all regions suggests possible contributions from carbonate and bicarbonate ions from runoff or atmospheric deposition.

### **Chemical Oxygen Demand (COD)**
The COD values vary significantly between regions. The Central region exhibits the highest mean COD in both seasons (19.79 mg/L dry, 19.12 mg/L wet), reflecting higher organic pollutant levels. Western region COD is notably lower (8.23 mg/L dry), indicating lesser organic contamination. Seasonal differences are subtle, but wet season values tend to decrease slightly, potentially due to dilution.

### **Biochemical Oxygen Demand (BOD)**
BOD, which measures the oxygen demand of biodegradable organic matter, remains consistently low across all regions, with means around 1.3–1.5 mg/L. The slight increase during wet seasons, especially in the Western region (1.45 mg/L), could signal increased organic matter from runoff. However, these values are within acceptable limits, suggesting minimal pollution from biodegradable organic sources.

### **General Observations**
Overall, the Central region exhibits higher ionic content, reflected in its conductivity, TDS, and COD values, possibly due to local geology or anthropogenic activities. The Western region generally shows lower values for most parameters, indicating relatively purer water. Seasonal patterns indicate a dilution effect during the wet season, with reduced ion concentrations and lower COD levels but slightly higher turbidity. The consistency in low turbidity and BOD values across all regions underscores the generally good quality of tap water, making it safe for consumption based on the parameters analyzed.

### Comparative Analysis: Seasons and Regions

```python
import pandas as pd
import scipy.stats as stats

# Separate data into dry and wet seasons for a parameter (e.g., pH)
dry = df[df['Season'] == 'Dry']
wet = df[df['Season'] == 'Wet']

# Paired t-test for normally distributed data (e.g., pH)
stat, p_value_ttest = stats.ttest_rel(dry['pH'], wet['pH'])
print(f'Paired t-test: stat={stat:.3f}, p-value={p_value_ttest:.3f}')

# Wilcoxon Signed-Rank Test for non-normally distributed data (e.g., TDS)
stat, p_value_wilcoxon = stats.wilcoxon(dry['TDS (ppm)'], wet['TDS (ppm)'])
print(f'Wilcoxon Signed-Rank Test: stat={stat:.3f}, p-value={p_value_wilcoxon:.3f}')
```

```python
# Define parameters to test
parameters = ['pH', 'TDS (ppm)', 'Tempt (°C)', 'COD (mg/L)', 'BOD (mg/L)',
              'Turbidity (NTU)', 'Alkalinity (mg/L)', 'Cond. (µ/cm)']

# Prepare a results DataFrame
results = pd.DataFrame(columns=['Variable',
                                 'Shapiro Stat', 'Shapiro p-val',
                                 'Anderson Stat', 'Anderson Normal?',
                                 'KS Stat', 'KS p-val',
                                 'Paired t-test Stat', 'Paired t-test p-val',
                                 'Wilcoxon Stat', 'Wilcoxon p-val'])

# Iterate through parameters
for param in parameters:
    # Get dry and wet season data
    dry_data = dry[param].dropna()
    wet_data = wet[param].dropna()

    # Perform normality tests
    shapiro_stat, shapiro_pval = stats.shapiro(dry_data)
    anderson_result = stats.anderson(dry_data)
    ks_stat, ks_pval = stats.kstest(dry_data, 'norm', args=(dry_data.mean(), dry_data.std()))

    # Check if Anderson-Darling test statistic indicates normality
    critical_values = anderson_result.critical_values
    significance_level = anderson_result.significance_level

    # Determine if the data is normal based on the critical value for the 5% significance level
    anderson_normal = any(anderson_result.statistic < cv for cv in critical_values)

    # Determine paired t-test or Wilcoxon test
    if shapiro_pval > 0.05:  # Normally distributed
        t_stat, t_pval = stats.ttest_rel(dry_data, wet_data)
        wilcoxon_stat, wilcoxon_pval = None, None  # Not applicable
    else:  # Not normally distributed
        t_stat, t_pval = None, None  # Not applicable
        wilcoxon_stat, wilcoxon_pval = stats.wilcoxon(dry_data, wet_data)

    # Create a new DataFrame for the results of the current parameter
    param_results = pd.DataFrame({
        'Variable': [param],
        'Shapiro Stat': [shapiro_stat],
        'Shapiro p-val': [shapiro_pval],
        'Anderson Stat': [anderson_result.statistic],
        'Anderson Normal?': [anderson_normal],  # True if normal, False otherwise
        'KS Stat': [ks_stat],
        'KS p-val': [ks_pval],
        'Paired t-test Stat': [t_stat],
        'Paired t-test p-val': [t_pval],
        'Wilcoxon Stat': [wilcoxon_stat],
        'Wilcoxon p-val': [wilcoxon_pval]
    })

    # Concatenate the results
    results = pd.concat([results, param_results], ignore_index=True)

# Display results
results
```

The analysis of normality and paired tests on tap water quality parameters provides insights into the distributional properties and potential group differences of the dataset. Below is a detailed discussion of the results.

### **Normality Tests**
The Shapiro-Wilk and Anderson-Darling tests provide assessments of normality for the parameters. The **pH**, **BOD**, **Turbidity**, and **Alkalinity** values fail the Shapiro-Wilk test (p-values < 0.05) and are confirmed as non-normally distributed by Anderson-Darling, where the "Normal?" column indicates "False." This implies significant deviations from a Gaussian distribution for these parameters, possibly due to outliers or intrinsic variability.

Conversely, parameters like **TDS**, **Temperature**, **COD**, and **Conductivity** pass both normality tests, as indicated by p-values > 0.05 in the Shapiro-Wilk test and "True" in the Anderson-Darling normality column. These parameters are likely symmetrically distributed, which may simplify subsequent parametric analyses.

### **Kolmogorov-Smirnov (KS) Test**
The KS test reinforces the normality results. Parameters like **pH**, **Turbidity**, and **BOD** show significant KS statistics with low p-values (< 0.05), confirming deviations from normality. For parameters such as **TDS**, **Temperature**, **COD**, and **Conductivity**, high p-values (> 0.05) indicate no significant deviation from the assumed distribution. This concordance between KS, Shapiro-Wilk, and Anderson-Darling tests strengthens the reliability of the findings.

### **Paired t-Test and Wilcoxon Test**
For parameters subjected to paired comparisons, the results highlight varying degrees of significance:
- **pH**: The Wilcoxon signed-rank test (p-value = 0.032) indicates a statistically significant difference between paired groups. However, no paired t-test is reported, likely due to non-normality.
- **BOD**: The Wilcoxon test does not find significance (p-value = 0.065), despite failing normality tests, indicating potential small sample sizes or low variability within the dataset.

```python
import pandas as pd
import numpy as np
from scipy import stats
from statsmodels.stats.multicomp import MultiComparison

# List of parameters to analyze
parameters = ['pH', 'TDS (ppm)', 'Tempt (°C)', 'COD (mg/L)', 'BOD (mg/L)',
              'Turbidity (NTU)', 'Alkalinity (mg/L)', 'Cond. (µ/cm)']

# Prepare results storage
results_list = []

for param in parameters:
    # Prepare data for Kruskal-Wallis Test
    group_data = [group[param].dropna() for name, group in df.groupby('Region')]

    # Perform Kruskal-Wallis Test
    kruskal_stat, p_val = stats.kruskal(*group_data)

    # Store the results in a dictionary
    results_list.append({
        'Variable': param,
        'Kruskal Stat': kruskal_stat,
        'p-value': p_val
    })

    # If Kruskal-Wallis is significant, perform Dunn's post-hoc test
    if p_val < 0.05:
        mc = MultiComparison(df[param], df['Region'])
        posthoc_results = mc.tukeyhsd()
        print(f"\nTukey's HSD results for {param}:\n", posthoc_results)

# Convert the results list to a DataFrame
kruskal_results = pd.DataFrame(results_list)

# Display Kruskal-Wallis results
print("\nKruskal-Wallis Results:")
kruskal_results
```

The Tukey's HSD and Kruskal-Wallis results provide complementary insights into differences among regional groups in water quality parameters. Here is a detailed discussion of the findings:

### **Tukey’s HSD Results**

The mean difference between the Central and Western regions is -1.0586 (p < 0.05), and between Greater Accra and Western regions is -0.8586 (p < 0.05). However, there is no significant difference between the Central and Greater Accra regions (p = 0.6031). These results suggest that the Western region has substantially lower pH levels compared to the other regions, highlighting potential issues with water acidity that may be linked to geological or anthropogenic factors in this area.

#### **COD (mg/L)**  
Significant differences are also observed in COD levels between the Western region and both the Central and Greater Accra regions. This indicates a higher presence of organic pollutants in water from the Western region. However, the difference between the Central and Greater Accra regions is not significant (p = 0.0705), suggesting relatively similar organic pollutant loads in these areas.

#### **Turbidity (NTU)**  
No statistically significant differences are observed in turbidity across the three regions. The p-values for all pairwise comparisons are greater than 0.05, although the difference between Greater Accra and the Western region approaches significance (p = 0.0561). These results indicate that turbidity levels are generally consistent across regions, implying similar particulate matter levels in water.

#### **Alkalinity (mg/L)**  
For alkalinity, significant differences are observed between the Central and Greater Accra regions (mean difference = 0.3929, p = 0.0175) and between the Greater Accra and Western regions (mean difference = -0.3836, p = 0.0207). These findings suggest variability in water buffering capacity, with Greater Accra having significantly different alkalinity levels compared to the other two regions.

### **Kruskal-Wallis Results**  
The Kruskal-Wallis test results corroborate Tukey’s HSD findings, particularly for parameters like pH and COD, which show significant differences across regions (p < 0.001). This non-parametric test further confirms the variability in water quality among regions. Specifically:
- **pH** (stat = 17.19, p = 0.0002): Significant regional differences align with the Tukey HSD analysis, emphasizing the distinct acidity levels in the Western region.

- **Turbidity** (stat = 6.06, p = 0.048): While Tukey's HSD indicated no significant pairwise differences, the Kruskal-Wallis test suggests marginal regional variation in turbidity.
- **Alkalinity** (stat = 6.89, p = 0.032): Differences in alkalinity are statistically significant, consistent with the Tukey's HSD analysis.

Parameters like **TDS**, **Temperature**, **BOD**, and **Conductivity** do not show significant differences across regions (p > 0.05), suggesting that these parameters are relatively stable regardless of location. This indicates uniformity in dissolved solids, temperature, biological oxygen demand, and electrical conductivity in the analyzed water samples.

### **Conclusion**  
The combined Tukey’s HSD and Kruskal-Wallis analyses reveal that water quality parameters such as pH, COD, and alkalinity exhibit significant spatial variation across regions, particularly with the Western region showing distinct deviations in acidity and organic pollution levels. On the other hand, parameters like turbidity exhibit marginal variability, while TDS, temperature, BOD, and conductivity are largely consistent across regions.
```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.metrics import silhouette_score

# Load your dataset
data = df.copy()   #pd.read_csv('your_data.csv')

# Select relevant features for clustering
features = data[['pH', 'Cond. (µ/cm)', 'TDS (ppm)', 'Tempt (°C)', 'Turbidity (NTU)', 'Alkalinity (mg/L)']]

# Normalize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Hierarchical Clustering
linked = linkage(scaled_features, method='ward')

plt.figure(figsize=(12, 8))
dendrogram(linked, orientation='top', labels=data['Region'].values, distance_sort='descending', show_leaf_counts=True)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

# Determine clusters using a threshold or number of clusters
threshold = 5  # Example threshold; adjust based on the dendrogram
clusters_hierarchical = fcluster(linked, threshold, criterion='distance')

# K-means Clustering
inertia = []
silhouette_scores = []
K = range(2, 11)  # Testing cluster sizes from 2 to 10

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_features)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(scaled_features, kmeans.labels_))

# Elbow method for K-means
plt.figure(figsize=(10, 6))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()

# Fit K-means with the chosen number of clusters (e.g., using elbow method results)
optimal_k = 3  # Replace with your determined optimal number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
kmeans.fit(scaled_features)
clusters_kmeans = kmeans.labels_

# Combine clusters with the original data
data['Hierarchical_Cluster'] = clusters_hierarchical
data['Kmeans_Cluster'] = clusters_kmeans

# Visualize K-means clusters
sns.scatterplot(data=data, x='pH', y='TDS (ppm)', hue='Kmeans_Cluster', palette='Set1')
plt.title('K-means Clustering (pH vs TDS)')
plt.show()

# Check for alignment with regions or seasonal differences
print("Cluster Distribution by Region:")
print(data.groupby(['Region', 'Kmeans_Cluster']).size())

print("\nCluster Distribution by Season:")
print(data.groupby(['Season', 'Kmeans_Cluster']).size())
```

In contrast, the Western and Greater Accra regions exhibit more heterogeneity, with samples forming distinct sub-clusters.
