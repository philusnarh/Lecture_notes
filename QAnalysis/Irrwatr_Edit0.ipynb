```python
import matplotlib.pyplot as plt
#%matplotlib inline
import numpy as np
import seaborn as sns
```

```python
plt.rc("text", usetex=False)
# plt.style.use('dark_background')
df_corr = df.drop('CONC(mg/L)', axis=1).corr() #data[relation].corr()
fig, ax = plt.subplots(figsize=(15, 12), dpi=300)

#mask
mask = np.triu(np.ones_like(df_corr, dtype=np.bool))
cmap = 'Spectral'

# plot heatmap
heatmap = sns.heatmap(df_corr,
                      mask=mask,
                      annot=True,
                      fmt=".2f",
                      cmap=cmap, # 'cubehelix'
                      vmin=-1, vmax=1,
                      cbar_kws={'shrink':.9, 'orientation': 'vertical'},
                      annot_kws={"size": 15})
# # Increase color bar font size
# heatmap.figure.colorbar(heatmap.collections[0], ax=heatmap, aspect=5, pad=0.02, shrink=0.7,
#                         cbar_kws={'orientation': 'vertical'}) #ticks=[your_ticks_here],
# cax = plt.gcf().axes[-1]
# cax.tick_params(labelsize=15)  # Set the font size for color bar

# Increase the color bar fontsize
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=15)  # Adjust the font size (12 in this example)



# yticks
plt.yticks(rotation=45, fontsize=15)
plt.xticks(rotation=70,fontsize=15)
# Hide grid lines
plt.grid(False)
```

```python
X_df = df.drop('CONC(mg/L)', axis=1)
y_df = df['CONC(mg/L)']

import random
from sklearn.ensemble import RandomForestRegressor as rfR
from yellowbrick.model_selection import FeatureImportances
plt.rcParams.update({
   'figure.figsize': (15.0, 10.0)
    })

c ='Nitrate Conc. in Irr. Water '
topN = 8
# val = randint(0, topN)

import matplotlib._color_data as mcd

# Show the feature selection
model = rfR(n_estimators=10, random_state=16)

viz = FeatureImportances(model,
                         xlabel='Relevant importance for {}'.format(c),
                        #  xlabel_fontsize=16,
                        # topn=topN,
                        stack=False,
#                          size=[15,10],
                        colors=random.sample(population=list(mcd.CSS4_COLORS.keys()), k=topN),
                        #  title="Feature Importances",
                        #  fontsize=25
                         )
viz.fit(X_df, y_df)

# # Increase x-label font sizes
# viz.set_xlabel('Relevant importance for {}'.format(c), label_size=16)  # Adjust label_size as needed

# # Increase y-label font sizes
# viz.set_ylabel("physicochemical parameters", label_size=16)  # Adjust label_size as needed


# Increase tick font sizes
viz.ax.tick_params(axis='both', which='major', labelsize=14)
viz.ax.tick_params(axis='both', which='minor', labelsize=12)

# Increase y-label font size
plt.xlabel('Relevant importance for {}'.format(c), fontsize=18)  # Adjust the font size as needed

# viz.title_fontsize = 20  # Adjust the title font size

# Finalize and display the visualizer
viz.show()
```
